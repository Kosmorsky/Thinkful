## 3. Now, check if your model meets the Gauss-Markov Conditions above. If some of the assumptions are not met, discuss the implications of the violations for the correctness of your model.

### Assumption 1: linearity of the model in its coefficients

```python
#pairplot
#heatmap

# drop unused columns
df2 = df.copy()
df2 = df2.drop(['apparenttemperature'], axis=1)
df2 = df2.drop(['visibility'], axis=1)
df2 = df2.drop(['loudcover'], axis=1)

sns.pairplot(df2)
```

### Assumption 2: the error term should be zero on average

```python
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)

predictions = lrm.predict(X)
errors = Y - predictions

print("Mean of the errors in the medical costs model is: {}".format(np.mean(errors)))
```

### Assumption 3: homoscedasticity

Error average is effectively zero

```python
plt.scatter(predictions, errors)
plt.xlabel('Predicted')
plt.ylabel('Residual')
plt.axhline(y=0)
plt.title('Residual vs. Predicted')
plt.show()

#deal with outliers
```

It seems we do not suffer from homoscedasticity. Let's investigate further...

```python
bart_stats = bartlett(predictions, errors)
lev_stats = levene(predictions, errors)

print("Bartlett test statistic value is {0:3g} and p value is {1:.3g}".format(bart_stats[0], bart_stats[1]))
print("Levene test statistic value is {0:3g} and p value is {1:.3g}".format(lev_stats[0], lev_stats[1]))
```

The p-values of both tests are lower than 0.05. So, the test results reject the null hypothesis which means our errors are heteroscedastic.

### Assumption 4: low multicollinearity

```python
plt.figure(figsize=(10,10))

sns.heatmap(df2.corr(), square=True, annot=True, linewidths=.5)

plt.show()
```

There are no significant correlations between features.

### Assumption 5: error terms should be uncorrelated with one another

```python
plt.plot(errors)
plt.show()
```

It seems that the error terms of our model are uncorrelated with each other.

```python
acf_data = acf(errors)

plt.plot(acf_data[1:])
plt.show()
```

### Assumption 6: features shouldn't be correlated with the errors

```python
plt.hist(errors)
plt.xlabel("errors")
plt.title("Histogram of the errors")
```

The errors look normally distributed.

```python
jb_stats = jarque_bera(errors)
norm_stats = normaltest(errors)

print("Jarque-Bera test statistics is {0} and p value is {1}".format(jb_stats[0], jb_stats[1]))
print("Normality test statistics is {0} and p value is {1}".format(norm_stats[0], norm_stats[1]))
```
