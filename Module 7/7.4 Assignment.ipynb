```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xlrd
%matplotlib inline
```

```python
df = pd.read_excel('tex_deathrow.xlsx')
```

# Age Distribution

```python
#plot 1: age distribution
age_dist = df['Age']

plt.hist(age_dist, color='red')
plt.title('Age Distribution: Texas Death Row since 1982')
plt.xlabel('Age')
plt.ylabel('# Executed')

plt.show()

print('Minimum Age: {}'.format(df['Age'].min()))
print('Maximum Age: {}'.format(df['Age'].max()))
```

This visualization shows the distribution of the ages of individuals who were executed, along with the minimum and maximum age. 
The data shows that those who were executed were most commonly between the ages of 30 and 40.
The youngest to be executed was 24, and after age 40, executions became consistently less common until the max age of 70.
Some of my speculations based on this data are:
1. Offenders under the age of 30 are likely seen as too young to be executed & garner sympathy and/or they commit less serious crimes
2. The older offenders are, the less incentive there is to execute them since they have less life to live anyway and/or old people are less likely to commit serious crimes (or the ones who did have already been executed)

# Race Breakdown

```python
race = df['Race']

#print pie chart
plt.subplot(1, 2, 1)
race.value_counts().plot.pie()

#print bar chart
plt.subplot(1, 2, 2)
race.value_counts().plot.bar(color='Purple')

plt.tight_layout()
plt.show()

#total people
num_total = len(df.index) - 1

#num & pct white
num_white = race.str.count("White").sum()
pct_white = num_white/num_total

#num & pct black
num_black = race.str.count("Black").sum()
pct_black = num_black/num_total

#num & pct hispanic
num_hispanic = race.str.count("Hispanic").sum()
pct_hispanic = num_hispanic/num_total

#num & pct other
num_other = race.str.count("Other").sum()
pct_other = num_other/num_total

print('{} people have been executed in Texas since 1982.'.format(num_total))
print('{} of the people executed were white, or {:.2%} of the total.'.format(num_white, pct_white))
print('{} of the people executed were black, or {:.2%} of the total.'.format(num_black, pct_black))
print('{} of the people executed were hispanic, or {:.2%} of the total.'.format(num_hispanic, pct_hispanic))
print('{} of the people executed were another race, or {:.2%} of the total.'.format(num_other, pct_other))

#note: couldn't figure out how to put numbers & percentages directly on pie chart
```

These visualizations show the breakdown of the races of the people executed.
The adition of the bar chart is to accurately show the numbers of individuals executed.
It is interesting to see the percentage of each race that was executed, although these figures are inconclusive without knowing the population of each race in Texas. (further research shows that 70% of the population is white & 12% is black, meaning there is a disproportionate abount of black people executed. data on hispanic population was unclear)
